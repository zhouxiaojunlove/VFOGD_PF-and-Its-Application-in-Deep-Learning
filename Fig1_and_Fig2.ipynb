```python
#f(x)=(1+x)**2
import numpy as np
import torch as t
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from d2l import torch as d2l
import torch.nn as nn
import time
```

```python
#import matplotlib   
#matplotlib.use('Agg')
```

```python
def f(x):
    return (x+1)**2

def x_grad_function(x,r):
    return math.gamma(3)*(x+1)**(2-r)/math.gamma(3-r)

def x_grad_function_fabs(x,r):
    return math.gamma(3)*(np.fabs(x+1))**(2-r)/math.gamma(3-r)
```

```python
range_size = 15
def fogd_pf(lr, xf_fabs):
    x = 10.0
    results = [x]
    for i in range(range_size):
        x -= lr * xf_fabs(x,alpha)
        #print('no')
        results.append(float(x))
    print(f'epoch, x: {x:f}')
    return results
```

```python
lr = 0.3     #learning rate
alpha = 0.7   #fractional order

timestart = time.time()
resultf0_7 = fogd_pf(lr, x_grad_function_fabs)
timeend = time.time()
print('time:',timeend-timestart)

alpha = 0.9   #fractional order
resultf0_9 = fogd_pf(lr, x_grad_function_fabs)

alpha = 1.1   #fractional order
resultf1_1 = fogd_pf(lr, x_grad_function_fabs)

alpha = 1.3   #fractional order
resultf1_3 = fogd_pf(lr, x_grad_function_fabs)
```

```python
fig = plt.figure(figsize=(10, 6),dpi=300)
x_major_locator=MultipleLocator(1)

#Fig1,alpha=0.7
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(resultf0_7,'b-.o',label=r'$\alpha  = 0.7$')
ax1.axhline(-1,linestyle='--',color='red')
ax1.set_xlim(0,15)
ax1.set_ylim(-10,10)
ax1.legend()
ax1.set_xlabel(r'$iteration$')
ax1.set_ylabel(r'$\theta $')
ax1.xaxis.set_major_locator(x_major_locator)
#ax1.yaxis.set_major_locator(x_major_locator)
ax1.annotate(r'${\theta _{\min }} =  - 1$', xy=(4,-1), xytext=(5,4),arrowprops=dict(facecolor='red'))

#Fig1,alpha=0.9
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(resultf0_9,'b-.o',label=r'$\alpha  = 0.9$')
ax2.axhline(-1,linestyle='--',color='red')
ax2.set_xlim(0,15)
ax2.set_ylim(-2,10)
ax2.legend()
ax2.set_xlabel(r'$iteration$')
ax2.set_ylabel(r'$\theta $')
ax2.xaxis.set_major_locator(x_major_locator)
#ax1.yaxis.set_major_locator(x_major_lo1
ax2.annotate(r'${\theta _{\min }} =  - 1$', xy=(1,-1), xytext=(4,2),arrowprops=dict(facecolor='red'))

#Fig1,alpha=1.1
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(resultf1_1,'b-.o',label=r'$\alpha  = 1.1$')
ax3.axhline(-1,linestyle='--',color='red')
ax3.set_xlim(0,15)
ax3.set_ylim(-2,10)
ax3.legend()
ax3.set_xlabel(r'$iteration$')
ax3.set_ylabel(r'$\theta $')
ax3.xaxis.set_major_locator(x_major_locator)
#ax1.yaxis.set_major_locator(x_major_locator)
ax3.annotate(r'${\theta _{\min }} =  - 1$', xy=(2,-1), xytext=(4,2),arrowprops=dict(facecolor='red'))

#Fig1,alpha=1.3
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(resultf1_3,'b-.o',label=r'$\alpha  = 1.3$')
ax4.axhline(-1,linestyle='--',color='red')
ax4.set_xlim(0,15)
ax4.set_ylim(-2,10)
ax4.legend()
ax4.set_xlabel(r'$iteration$')
ax4.set_ylabel(r'$\theta $')
ax4.xaxis.set_major_locator(x_major_locator)
#ax1.yaxis.set_major_locator(x_major_locator)
ax4.annotate(r'${\theta _{\min }} =  - 1$', xy=(2,-1), xytext=(5,4),arrowprops=dict(facecolor='red'))

plt.savefig('picture/Fig1.eps',bbox_inches='tight',format='eps')
plt.savefig('picture/Fig1.svg',bbox_inches='tight',format='svg')
plt.savefig('picture/Fig1.pdf',bbox_inches='tight',format='pdf')
plt.show()
```
