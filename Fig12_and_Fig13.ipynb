```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
#sgd
vfosgd_pf_loss = pd.read_csv('fashion_mnist_loss_accuracy/VFOGD_PF_Loss.csv')
vfosgd_pf_accuracy = pd.read_csv('fashion_mnist_loss_accuracy/VFOGD_PF_Accuracy.csv')
#adam
vfoadam_pf_loss = pd.read_csv('fashion_mnist_loss_accuracy/VFOAdam_PF_Loss.csv')
vfoadam_pf_accuracy = pd.read_csv('fashion_mnist_loss_accuracy/VFOAdam_PF_Accuracy.csv')
```

```python
fig = plt.figure(figsize=(10,6),dpi=300)
plt.rcParams.update({'font.size': 8})
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(vfosgd_pf_loss.iloc[0,1:150],'r--',label=r'$\alpha=0.2$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[1,1:150],'b--',label=r'$\alpha=0.4$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[2,1:150],'y--',label=r'$\alpha=0.6$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[3,1:150],'g--',label=r'$\alpha=0.8$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[4,1:150],'c--',label=r'$\alpha=1.0$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[5,1:150],'k-',label=r'$\alpha=1.2$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[6,1:150],'m-',label=r'$\alpha=1.4$',markersize='1',linewidth ='1')
ax1=plt.gca()
ax1.xaxis.set_major_locator(MultipleLocator(20))
ax1.set_xlim(-1,150)
ax1.set_ylim(0.05,0.7)
#ax1.set_xlabel(r'$epoch$')
ax1.set_ylabel(r'$loss$')
ax1.set_title(r'VFOSGD_PF_Loss',bbox=dict(facecolor='y', edgecolor='blue', alpha=0.65 ),x=0.5,y=0.7)
ax1.legend()
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(vfosgd_pf_accuracy.iloc[0,1:150],'r--',label=r'$\alpha=0.2$',markersize='1',linewidth ='1')
ax2.plot(vfosgd_pf_accuracy.iloc[1,1:150],'b--',label=r'$\alpha=0.4$',markersize='1',linewidth ='1')
ax2.plot(vfosgd_pf_accuracy.iloc[2,1:150],'y--',label=r'$\alpha=0.6$',markersize='1',linewidth ='1')
ax2.plot(vfosgd_pf_accuracy.iloc[3,1:150],'g--',label=r'$\alpha=0.8$',markersize='1',linewidth ='1')
ax2.plot(vfosgd_pf_accuracy.iloc[4,1:150],'c--',label=r'$\alpha=1.0$',markersize='1',linewidth ='1')
ax2.plot(vfosgd_pf_accuracy.iloc[5,1:150],'k-',label=r'$\alpha=1.2$',markersize='1',linewidth ='1')
ax2.plot(vfosgd_pf_accuracy.iloc[6,1:150],'m-',label=r'$\alpha=1.4$',markersize='1',linewidth ='1')
ax2=plt.gca()
ax2.xaxis.set_major_locator(MultipleLocator(20))
ax2.set_xlim(-1,150)
ax2.set_ylim(0.75,0.91)
ax2.set_xlabel(r'$epoch$')
ax2.set_ylabel(r'$accuracy$')
ax2.set_title(r'VFOSGD_PF_Accuracy',bbox=dict(facecolor='y', edgecolor='blue', alpha=0.65 ),x=0.5,y=0.3)
ax2.legend()
plt.savefig('picture/Fig12.eps',bbox_inches='tight',format='eps')
plt.savefig('picture/Fig12.svg',bbox_inches='tight',format='svg')
plt.savefig('picture/Fig12.pdf',bbox_inches='tight',format='pdf')
plt.show()
```
