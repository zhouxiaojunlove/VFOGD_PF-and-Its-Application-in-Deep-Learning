```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
```

```python
vfosgd_pf_loss = pd.read_csv('DJIA_and_ETTh1_loss_accuracy/DJIA_VFOGD_PF_Loss_MLP.csv')
vfoadam_pf_loss = pd.read_csv('DJIA_and_ETTh1_loss_accuracy/DJIA_VFOAdam_PF_Loss_MLP.csv')
```

```python
vfosgd_pf_loss
```

```python
fig = plt.figure(figsize=(10,6),dpi=300)
plt.rcParams.update({'font.size': 8})
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(vfosgd_pf_loss.iloc[0,1:500],'r--',label=r'$\alpha=0.6$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[1,1:500],'b--',label=r'$\alpha=0.8$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[2,1:500],'y--',label=r'$\alpha=1.0$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[3,1:500],'g-',label=r'$\alpha=1.2$',markersize='1',linewidth ='1')
ax1=plt.gca()
ax1.xaxis.set_major_locator(MultipleLocator(100))
ax1.set_xlim(0,500)
ax1.set_ylim(0.004,0.02)
ax1.set_ylabel(r'$loss$')
ax1.set_title(r'VFOSGD_PF',bbox=dict(facecolor='y', edgecolor='blue', alpha=0.65 ),x=0.5,y=0.7)
ax1.legend()
ax3 = fig.add_subplot(2, 1, 2)
ax3.plot(vfoadam_pf_loss.iloc[0,1:300],'r--',label=r'$\alpha=0.6$',markersize='1',linewidth ='1')
ax3.plot(vfoadam_pf_loss.iloc[1,1:300],'b--',label=r'$\alpha=0.8$',markersize='1',linewidth ='1')
ax3.plot(vfoadam_pf_loss.iloc[2,1:300],'y--',label=r'$\alpha=1.0$',markersize='1',linewidth ='1')
ax3.plot(vfoadam_pf_loss.iloc[3,1:300],'g-',label=r'$\alpha=1.2$',markersize='1',linewidth ='1')
ax3.plot(vfoadam_pf_loss.iloc[4,1:300],'c--',label=r'$\alpha=1.4$',markersize='1',linewidth ='1')
ax3=plt.gca()
ax3.xaxis.set_major_locator(MultipleLocator(50))
ax3.set_xlim(40,300)
ax3.set_ylim(0.003,0.02)
ax3.set_xlabel(r'$epoch$')
ax3.set_ylabel(r'$loss$')
ax3.set_title('VFOAdam_PF',bbox=dict(facecolor='y', edgecolor='blue', alpha=0.65 ),x=0.5,y=0.7)
ax3.legend()
plt.savefig('picture/Fig16.eps',bbox_inches='tight',format='eps')
plt.savefig('picture/Fig16.svg',bbox_inches='tight',format='svg')
plt.savefig('picture/Fig16.pdf',bbox_inches='tight',format='pdf')
plt.show()
```
