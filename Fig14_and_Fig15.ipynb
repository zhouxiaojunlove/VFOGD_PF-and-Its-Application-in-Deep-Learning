```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
#sgd and adam at DJIA
```

```python
vfosgd_pf_loss = pd.read_csv('DJIA_and_ETTh1_loss_accuracy/DJIA_VFOGD_PF_Loss.csv')
vfoadam_pf_loss = pd.read_csv('DJIA_and_ETTh1_loss_accuracy/DJIA_VFOAdam_PF_Loss.csv')
```

```python
fig = plt.figure(figsize=(10,6),dpi=300)
plt.rcParams.update({'font.size': 8})
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(vfosgd_pf_loss.iloc[0,1:1000],'r--',label=r'$\alpha=0.6$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[1,1:1000],'b--',label=r'$\alpha=0.8$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[2,1:1000],'y--',label=r'$\alpha=1.0$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[3,1:1000],'g--',label=r'$\alpha=1.2$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[4,1:1000],'c--',label=r'$\alpha=1.25$',markersize='1',linewidth ='1')
ax1.plot(vfosgd_pf_loss.iloc[5,1:1000],'k-',label=r'$\alpha=1.28$',markersize='1',linewidth ='1')
ax1=plt.gca()
ax1.xaxis.set_major_locator(MultipleLocator(100))
ax1.set_xlim(0,1000)
ax1.set_ylim(0.02,0.16)
ax1.set_ylabel(r'$loss$')
ax1.set_title(r'VFOSGD_PF',bbox=dict(facecolor='y', edgecolor='blue', alpha=0.65 ),x=0.5,y=0.5)
ax1.legend()
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(vfoadam_pf_loss.iloc[0,1:100],'r--',label=r'$\alpha=0.6$',markersize='1',linewidth ='1')
ax2.plot(vfoadam_pf_loss.iloc[1,1:100],'b--',label=r'$\alpha=0.8$',markersize='1',linewidth ='1')
ax2.plot(vfoadam_pf_loss.iloc[2,1:100],'y-',label=r'$\alpha=1.0$',markersize='1',linewidth ='1')
ax2.plot(vfoadam_pf_loss.iloc[3,1:100],'g--',label=r'$\alpha=1.2$',markersize='1',linewidth ='1')
ax2.plot(vfoadam_pf_loss.iloc[4,1:100],'c--',label=r'$\alpha=1.4$',markersize='1',linewidth ='1')
ax2.plot(vfoadam_pf_loss.iloc[5,1:100],'k-',label=r'$\alpha=1.6$',markersize='1',linewidth ='1')
ax2.plot(vfoadam_pf_loss.iloc[6,1:100],'k--',label=r'$\alpha=1.8$',markersize='1',linewidth ='1')
ax2=plt.gca()
ax2.xaxis.set_major_locator(MultipleLocator(10))
ax2.set_xlim(0,100)
ax2.set_ylim(0.05,0.40)
ax2.set_xlabel(r'$epoch$')
ax2.set_ylabel(r'$loss$')
ax2.set_title(r'VFOAdam_PF',bbox=dict(facecolor='y', edgecolor='blue', alpha=0.65 ),x=0.5,y=0.5)
ax2.legend()
plt.savefig('picture/Fig14.eps',bbox_inches='tight',format='eps')
plt.savefig('picture/Fig14.svg',bbox_inches='tight',format='svg')
plt.savefig('picture/Fig14.pdf',bbox_inches='tight',format='pdf')
plt.show()
```
